```python
import PySimpleGUI as sg
from mysql import connector
 
def connect_to_sql(p):
    sqlcon = connector.connect(
    host = "localhost",
    user = "root",
    password = p)
    
    return sqlcon

condb = [[sg.Text("Enter Database Password:")],
         [sg.InputText(password_char= "*"),sg.Button("Connect")],
        [sg.Text("",key="con",size = (30,1))]]

buttons = [[sg.Button('List of Databases')],
           [sg.InputText(),sg.Button('Query from User')],[sg.Text("")],[sg.Text("Queries of Bank Database",size = (45,1), justification = 'center',font='bold')],
           [sg.Button("Departments and Employees",size=(30,1)),sg.Button("Total Accounts",size=(30,1))],
           [sg.Button("Customer and Available Balance",size=(30,1)),sg.Button("Certificate of Deposit Products",size=(30,1))],
           [sg.Button("Roles of Employees",size=(30,2)), sg.Button("Officers Allotted to Customers and Their Business",size=(30,2))]
          ]

output = [[sg.Text("Result", size=(45,1),justification = 'center',font = "bold")],[sg.Text("",size=(50,10),key="result")]]

layout = [[sg.Column(condb)],
    [sg.Text('_'*70)],
    [sg.Column(buttons)], 
[sg.Text('_'*70)],
[sg.Column(output)]]

window = sg.Window("Bank Gui", layout)

while True:
    event, values = window.read()
  
    if event == "EXIT" or event == sg.WIN_CLOSED:
        break
        
    if event:
        if event == "Connect":
            sqlcon = connect_to_sql(values[0])
            window["con"].Update("Connected Successfully")

        if event == "List of Databases":
                q = "Show Databases;"
                cur = sqlcon.cursor()
                cur.execute(q)

                values = []
                for x in cur:
                    values.append(list(x))
                    
                out = ""
                for i in values:
                    out = out + str(i) + "\n"

                window["result"].Update(out)

                #window["result"].Update(values)
                
        elif event == "Query from User":
            q = values[1]
            cur = sqlcon.cursor()
            cur.execute(q)

            values = []
            for x in cur:
                 values.append(x)

            window["result"].Update(values)

        
        elif event == "Departments and Employees":
            q1 = "use bank;"
            q = "select d.name, count(e.DEPT_ID) as 'number of employees' from employee e inner join department d on e.DEPT_ID = d.DEPT_ID group by e.DEPT_ID;"
            cur = sqlcon.cursor()
            cur.execute(q1)
            cur.execute(q)

            values = []
            for x in cur:
                 values.append(list(x))

            out = ""
            for i in values:
                out = out + str(i) + "\n"

            window["result"].Update(out)
        
        elif event == "Customer and Available Balance":
            q1 = "use bank;"
            q = "select a.CUST_ID,concat(i.first_name,' ', i.last_name) as 'customer name',sum(a.avail_balance) as 'available balance' from account a inner join individual i on i.CUST_ID = a.CUST_ID inner join customer c on c.CUST_ID = a.CUST_ID group by CUST_ID;"
            cur = sqlcon.cursor()
            cur.execute(q1)
            cur.execute(q)

            values = []
            for x in cur:
                 values.append(list(x))
            
            out = ""
            for i in values:
                out = out + str(i) + "\n"

            window["result"].Update(out)
        
        elif event == "Total Accounts":
            q1 = "use bank;"
            q = " select count(account_id) as 'Total number of accounts' from account;"
            cur = sqlcon.cursor()
            cur.execute(q1)
            cur.execute(q)

            for x in cur:
                window["result"].Update(list(x))
            
        elif event == "Certificate of Deposit Products":
            q1 = "use bank;"
            q = "select PRODUCT_CD,count(product_cd) as 'number of CD products' from account group by PRODUCT_CD order by count(PRODUCT_CD) desc;"
            cur = sqlcon.cursor()
            cur.execute(q1)
            cur.execute(q)

            values = []
            for x in cur:
                 values.append(list(x))
            
            out = ""
            for i in values:
                out = out + str(i) + "\n"

            window["result"].Update(out)

        elif event == "Roles of Employees":
            q1 = "use bank;"
            q = "select Title,count(TITLE) as 'number of employees' from employee group by TITLE order by count(Title) desc;"
            cur = sqlcon.cursor()
            cur.execute(q1)
            cur.execute(q)

            values = []
            for x in cur:
                 values.append(list(x))
            
            out = ""
            for i in values:
                out = out + str(i) + "\n"

            window["result"].Update(out)
            
           

        elif event == "Officers Allotted to Customers and Their Business":
            q1 = "use bank;"
            q = " select concat(o.first_name,' ', o.last_name) as 'officer name',o.cust_id,b.NAME as 'customer business' from business b inner join officer o on b.CUST_ID = o.CUST_ID;"
            cur = sqlcon.cursor()
            cur.execute(q1)
            cur.execute(q)

            values = []
            for x in cur:
                 values.append(list(x))
            
            out = ""
            for i in values:
                out = out + str(i) + "\n"

            window["result"].Update(out)
            
window.close()
```
